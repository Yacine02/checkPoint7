```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier #Importing Random Forest Classifier
from sklearn import metrics
import graphviz
```

```python
df = pd.read_csv('titanic-passengers.csv', sep=';') #retrieving data from titanic-passengers.csv in a pandas dataframe 
titanic_copy = df.copy() #make a copy
```

# Prétaitement

# Nettoyage

```python
df.head()#display the header of my dataframe
```

```python
df.isnull().sum()#display the number of missing data in each column
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True )#replace missing values in the Age column with the mean the mode
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True )#replace missing values in the Cabin column with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True )#replace missing values in the Embarked column with
```

```python
df.isnull().sum()
```

# Transformation

```python
Survivived_dict = { "Yes":1, "No":0}#Replace categorical values Yes and No with numeric values 1 and 0
df['Survived'].replace(Survivived_dict, inplace=True)#insert it in the dataframe
```

```python
sex_dict = { "male":1, "female":0}#Replace categorical values male and female with numeric values 1 and 0
df['Sex'].replace(sex_dict, inplace=True)#insert it in the dataframe
df.head()
```

# Selection des fonctionnalités

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )#display the correlation between my data
```

```python
df.drop('PassengerId', axis=1, inplace=True)#drop the PassengerId column
df.drop('Name', axis=1, inplace=True)#drop the Name column
df.drop('SibSp', axis=1, inplace=True)#drop the SibSp column
df.drop('Parch', axis=1, inplace=True)#drop the Parch column
df.drop('Ticket', axis=1, inplace=True)#drop the Ticket column
df.drop('Fare', axis=1, inplace=True)#drop the Fare column
df.drop('Cabin', axis=1, inplace=True)#drop the Cabin column
df.drop('Embarked', axis=1, inplace=True)#drop the Embarked column
```

```python
df.head()
```

# Appliquons l'arbre de décision

```python
#features extraction
x = df[['Pclass', 'Sex','Age']]
y = df['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=10)#splitting data with test size of 20%
```

```python
#applying tree algorithm
classifier = tree.DecisionTreeClassifier()  
classifier.fit(x_train, y_train)   #fitting our model
y_pred=classifier.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

# Tracé de l'arbre de décision

```python
dot_data =tree.export_graphviz(classifier, out_file =None, filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("data")
graph
```

# Prediction manuel

 On peut dire ici que la meilleure répartition est le le sexe du passager car il a la division qui coûte le moins. 

# Modificqtion des paramètres de l'arbre de décision (au moins deux paramètres)

```python
dot_data = tree.export_graphviz(classifier, out_file=None, 
                                feature_names = x.columns,
                                max_depth = 3,
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

# Random forest

```python
clf=RandomForestClassifier(n_estimators=10)  #Creating a random forest with 100 decision trees
clf.fit(x_train, y_train)  #Training our model
y_pred=clf.predict(x_test)  #testing our model
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))  #Measuring the accuracy of our model
```

Accuracy of the random forest is greater than the accuracy of the decision tree.
